# Data preprocessing


```python
# Read data
import pandas as pd
df = pd.read_excel('./audiogram_data.xlsx',sheet_name=0)
```

```python
df_stripped = df[1:] # remove first row (feature names)

df_stripped = df_stripped.drop(df_stripped.columns[0], axis=1) # remove first column (row numbers)
df_stripped = df_stripped.drop(df_stripped.columns[4], axis=1) # remove device type column 
df_stripped = df_stripped.drop("Unnamed: 22",axis =1) # remove empty column

adg_data = df_stripped
df_stripped.iloc[:,3:] 
```

```python
age = (df_stripped["Operation year"] - df_stripped["Birth year"]).astype(int) # Compute age 
adg_data.insert(loc=1, column='Age', value=age) # insert age column
adg_data = adg_data.drop(["Operation year","Birth year"],axis=1) # Drop operation year and birth year columns
adg_data.iloc[:,2:] 
```

```python
# Adding other constructed features
air_means = (df_stripped.iloc[:,5:15]).mean(axis=1)
bone_means = (df_stripped.iloc[:,15:21]).mean(axis=1)
air_bone_ratio = bone_means / air_means
air_bone_mp = bone_means * air_means

pair_meana1 =  (df_stripped['125BefAir'] + df_stripped['250BefAir'] ) / 2
pair_meana2 =  (df_stripped['500BefAir'] + df_stripped['1000BefAir'] ) / 2
pair_meana3 =  (df_stripped['1500BefAir'] + df_stripped['2000BefAir'] ) / 2
pair_meana4 =  (df_stripped['3000BefAir'] + df_stripped['4000BefAir'] ) / 2
pair_meana5 =  (df_stripped['6000BefAir'] + df_stripped['8000BefAir'] ) / 2
pair_meanb1 =  (df_stripped['500BefBone'] + df_stripped['1000BefBone'] ) / 2
pair_meanb2 =  (df_stripped['1500BefBone'] + df_stripped['2000BefBone'] ) / 2
pair_meanb3 =  (df_stripped['3000BefBone'] + df_stripped['4000BefBone'] ) / 2

pta_air = (df_stripped.iloc[:,[7,8,10,11]]).mean(axis=1)
pta_bone = (df_stripped.iloc[:,[15,16,18,19]]).mean(axis=1)
abg = pta_air - pta_bone


adg_data.insert(loc=3, column='Air mean', value=air_means)
adg_data.insert(loc=4, column='Bone mean', value=bone_means)
adg_data.insert(loc=5, column='Bone/Air', value=air_bone_ratio)
adg_data.insert(loc=6, column='Bone*Air', value=air_bone_mp)

adg_data.insert(loc=7, column='125_250BA', value=pair_meana1)
adg_data.insert(loc=8, column='500_1000BA', value=pair_meana2)
adg_data.insert(loc=9, column='1500_2000BA', value=pair_meana3)
adg_data.insert(loc=10, column='3000_4000BA', value=pair_meana4)
adg_data.insert(loc=11, column='6000_8000BA', value=pair_meana5)
adg_data.insert(loc=12, column='500_1000BB', value=pair_meanb1)
adg_data.insert(loc=13, column='1500_2000BB', value=pair_meanb2)
adg_data.insert(loc=14, column='3000_4000BB', value=pair_meanb3)

adg_data.insert(loc=15, column='ABG', value=abg)



adg_data
```

```python
print(sum(adg_data.isna().sum())) # Check for missing values
print(adg_data['Gender-0=M, 1 =F'].value_counts()) # gender distribution
print(adg_data['Side 0=L, 1=R'].value_counts()) # ear side distribution

## Save data here
adg_data.to_csv('variables/processed.csv',index=False)
```
